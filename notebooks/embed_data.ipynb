```python
BASE_DIR = "models"
PHOBERT_DIR = f"{BASE_DIR}/phobert"
EMBEDDING_DIR = f"{BASE_DIR}/cache_embeddings"

TRAIN_PATH = "my_data\train_remap.csv"
DEV_PATH = "my_data\dev.csv"
TEST_PATH = "my_data\test.csv"

LABEL_MAP_PATH = "src\label_map.json"

print('oke')
```

```python
train_texts = train_df["title"].astype(str).tolist()
train_labels = train_df["label"].astype(int).tolist()

dev_texts = dev_df["title"].astype(str).tolist()
dev_labels = dev_df["label"].astype(int).tolist()

test_texts = test_df["title"].astype(str).tolist()
test_labels = test_df["label"].astype(int).tolist()

print(len(train_texts), len(train_labels))
print(len(dev_texts), len(dev_labels))
print(len(test_texts), len(test_labels))
```

```python
tokenizer, phobert_model = load_phobert(PHOBERT_DIR)
```

```python
X_train, y_train = get_or_create_embeddings(
    texts=train_texts,
    labels=train_labels,
    embed_fn=embed_texts,
    tokenizer=tokenizer,
    model=phobert_model,
    save_dir=EMBEDDING_DIR,
    split_name="train"
)

X_dev, y_dev = get_or_create_embeddings(
    texts=dev_texts,
    labels=dev_labels,
    embed_fn=embed_texts,
    tokenizer=tokenizer,
    model=phobert_model,
    save_dir=EMBEDDING_DIR,
    split_name="dev"
)

X_test, y_test = get_or_create_embeddings(
    texts=test_texts,
    labels=test_labels,
    embed_fn=embed_texts,
    tokenizer=tokenizer,
    model=phobert_model,
    save_dir=EMBEDDING_DIR,
    split_name="test"
)

print("Train:", X_train.shape, y_train.shape)
print("Dev  :", X_dev.shape, y_dev.shape)
print("Test :", X_test.shape, y_test.shape)
```
